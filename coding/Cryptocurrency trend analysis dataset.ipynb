## Data Cleaning

```python
### to check missing values
df.isnull().sum()
```

```python
## drop rows with missing values
df.dropna()
```

```python
# Drop rows with missing essential values or fill them
import pandas as pd
df.dropna(subset=["Coin Name", "Symbol"], inplace=True)
print(df.head())
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
print(df.tail())
```

```python
df.head()
```

```python
df.tail()
```

```python
# Clean and convert 'Market Cap' to numeric
df['Market Cap'] = pd.to_numeric(df['Market Cap'].astype(str).str.replace(r'[^\d.]', '', regex=True), errors='coerce')
```

## Data Pre-Processing

```python
## Convert percentage strings to float
percentage_cols = ["1h", "7d", "30d"]
for col in percentage_cols:
 df[col] = df[col].replace('%', '', regex=True)
print(df.head())
```

```python
# Convert supply columns to consistent numeric format
def convert_supply(supply):
    if 'Billion' in supply:
        return float(supply.replace(' Billion', '')) * 1e9
    elif 'Million' in supply:
        return float(supply.replace(' Million', '')) * 1e6
    else:
        try:
            return float(supply.replace(',', ''))
        except:
            return 0
print(convert_supply('1 Billion'))
print(convert_supply('2 Million'))
print(convert_supply('3,000'))
print(convert_supply('Invalid Input'))
```

```python
df['Total Supply'] = df['Total Supply'].astype(str).apply(convert_supply)
df['Circulating Supply'] = df['Circulating Supply'].astype(str).apply(convert_supply)
print(df[['Total Supply','Circulating Supply']])
```

```python
### top 10 cryptocurrencies by market cap
top10 = df.nlargest(10, 'Market Cap')
print(top10)
```

```python
## Barplot
plt.figure(figsize=(12, 6))
sns.barplot(x="Symbol", y="Market Cap", data=top10)
plt.title("Top 10 Cryptocurrencies by Market Cap")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
## histogram 
plt.figure(figsize=(10, 6))
sns.histplot(df['Market Cap'], bins=30, kde=True, color="skyblue")
plt.title("Histogram of Market Cap")
plt.show()
```

```python
### violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(y=df['24h'], color="lightgreen")
plt.title("Violin Plot of 24h Change")
plt.show()
```

```python
cols_to_clean1=df.select_dtypes(include='number').columns
cols_to_clean1
```

```python
### catplot
top_10 = df.nlargest(10, 'Market Cap')
sns.catplot(x='Symbol', y='Market Cap', kind='bar', data=top_10, palette="pastel", height=6, aspect=1.5)
plt.title("Top 10 Cryptos by Market Cap")
plt.show()
```

```python
### correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[cols_to_clean1].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
